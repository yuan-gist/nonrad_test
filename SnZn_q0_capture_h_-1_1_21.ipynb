```python
# use neutral structure and datapoints of disp_002,004,006
import matplotlib as mpl
import matplotlib.pyplot as plt
from nonrad.ccd import get_Q_from_struct
from nonrad.elphon import get_Wif_from_WSWQ
from pathlib import Path
from pymatgen.core.structure import Structure
from glob import glob

cc_dir = "/pscratch/sd/z/zkyuan/todos/kesterites/CZTS/defect/ZnCu2SnS4/static/nonrad" + \
         "/" + \
         "/as_1_Sn_on_Zn/q0_capture_h_-1_1_21/excited"

SnZn_4 = Structure.from_file(f"{cc_dir}/4/CONTCAR")
SnZn_5 = Structure.from_file(f"{cc_dir}/5/CONTCAR")
SnZn_6 = Structure.from_file(f"{cc_dir}/6/CONTCAR")


WSWQs = []
for d in glob(f"{cc_dir}/*"):
        if d[-2:] in ["/4", "/5", "/6"]:
            pd = Path(d)
            Q = get_Q_from_struct(SnZn_5, SnZn_6, f"{pd}/CONTCAR")
            path_wswq = f"{pd}/eph/WSWQ"
            WSWQs.append((Q, path_wswq))
print(WSWQs)      
fig = plt.figure(figsize=(12, 5))
ground_vr_path = f"{cc_dir}/5/vasprun.xml" 

Wifs = get_Wif_from_WSWQ(WSWQs, ground_vr_path, 2305, 2304, spin=1, fig=fig)

plt.tight_layout()
plt.ylim(-1e-6,1e-6)
plt.show()
```
